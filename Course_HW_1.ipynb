```python
# посмотрим датасет и переименуем столбцы для дальнейшего обединения в общий датасет 
df_students.rename(columns={'id' : 'student_id'}, inplace=True)
df_students.head()
```

```python
# удаляем столбец city и удаляем строки с пустыми значениями в колонке birthday
df_4.pop('city')
df_4.dropna()
```

```python
# с целью избавиться от ошибочных значений birthday оставляем диапозон перполагаемых верных значений
df_4 = df_4[(df_4['birthday'] > '1941-01-01') & (df_4['birthday'] < '2007-01-01')]
df_4
```

```python
now = pd.Timestamp('now') # создаем переменную с сегодняшей датой
df_4['birthday'] = pd.to_datetime(df_4['birthday'], format='%Y/%m/%d') # меняем тип колонки на datetime
df_4['age'] = (now - df_4['birthday']).astype('<m8[Y]') # создаем новую колонку с возрастом студентов
df_4
```

```python
print(df_4.age.min())
print(df_4.age.max())
print(df_4.age.mean())
print(df_4.age.median()) # минимальный, максимальный, средний, медианный возраст студентов 
```

```python
df_age = df.merge(df_4, on='student_id', how='inner') # соединяем датасеты для дльнейших расчетов по возрасту
df_age.head()
```

```python
df_g = df_age.groupby('course_title').agg({'age' : 'min'})# минимальный возраст на каждом курсе
df_g.sort_values(by='age', ascending=False)
```

```python
df_g = df_age.groupby('course_title').agg({'age' : 'max'})# максимальный возраст на каждом курсе
df_g.sort_values(by='age', ascending=False)
```

```python
df_g = df_age.groupby('course_title').agg({'age' : 'mean'})# средний возраст на каждом курсе
df_g.sort_values(by='age', ascending=False)
```

```python
df_g = df_age.groupby('course_title').agg({'age' : 'median'})# медианный возраст на каждом курсе
df_g.sort_values(by='age', ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(18, 10)) # bar-chart, отражающий количество студентов на каждом курсе 

sns.barplot(data=df_3, x = df_3.index, y='student_id', ax=ax)

ax.set_title('Количество студентов на каждом курсе', fontsize=18)
ax.set_xlabel('Название курса', fontsize=16)
ax.set_ylabel('Количество студентов', fontsize=14)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor", fontsize=13)
plt.show()
```

```python
# создаем датасет с количеством студентов на каждом курсе и  сфере курса
df_3q = df.groupby('course_title').agg({'student_id' : 'nunique', 'course_field' : 'min'})
df_3q = df_3q.sort_values(by='student_id', ascending=False)
df_3q
```

```python
fig, ax = plt.subplots(figsize=(16, 8)) # горизонтальный bar-chart, отражающий количество студентов на каждом курсе 

sns.barplot(data=df_3q, y = df_3q.index, x='student_id',
            orient ='h', hue='course_field', dodge=False,
            ax=ax, alpha=0.1)
ax.set_title('Количество студентов на каждом курсе', fontsize=18)
ax.set_ylabel('Название курса', fontsize=16)
ax.set_xlabel('Количество студентов', fontsize=16)
ax.set_xticks(list(range(0, 2500, 500)) +  [df_3q['student_id'].max()] + [df_3q['student_id'].median()])

ax.axvline(x=df_3q['student_id'].median(), color='green', linewidth=1, linestyle='--')
ax.legend(title='Сфера: ',title_fontsize=14,  fontsize=14)
plt.setp(ax.get_yticklabels(), fontsize=14)

plt.show()
```

```python
# создадим колонку student_start с единичными значениями(для подсчета общего количества начавших, начали все без исключения)
df_con['student_start'] = 1
```

```python
# создаем колонку student_finish так же с единичными значениями(за исключением у кого не сдана сдана домашняя работа,
# их мы находим и ставим 0 при помощи функции)
def finish_0(data):
    if data == '2020-01-01':
        return 0
    else:
        return 1
df_con['student_finish'] = df_con['finish_date'].apply(lambda x: finish_0(x))
```

```python
df_con
```

```python
# создаем датафрейм с суммой начатых и сданных домашних работ по каждому модулу
df_con_sum = df_con\
    .groupby(['course_title', 'module_number', 'module_title'])\
    .agg({'student_start' : np.sum, 'student_finish' : np.sum})\
    .reset_index()\
    .sort_values(by=['course_title', 'module_number'], ascending=False)
df_con_sum
```

```python
# создаем колонку с нулевыми значениями
df_con_sum['convers'] = 0.0
```

```python
# считаем конверсию перехода студентов из одного модуля в другой на каждом курсе
for x, y in enumerate(df_con_sum.iloc[:-1].index):
    if (df_con_sum['course_title'][y] != df_con_sum['course_title'][y-1]) or (df_con_sum['student_finish'][y-1] == 0):
        df_con_sum['convers'][y] = 0
    else:
        df_con_sum['convers'][y] = int((df_con_sum['student_start'][y] / df_con_sum['student_finish'][y-1]) * 100 )    
df_con_sum
```

## Обоснованивание метрики

За метрику успеваемости я решил взять процентиль скорости прохождения курса до последней завершенного модуля. Чей процентиль 0.95 и более попадают в список проблемных студентов.  <br>
Скорость прохождение курса рассчитывается от даты начала первой домашней работы до даты последней домашней работы в модуле. Так по каждому модулю выполненному студентом. 
Далее рассчитываем процентиль по каждому модулю на каждом курсе. После берем значение процентиля каждого студента того модуля, который он последний сдал. По этим значениям мы получаем информацию, является ли студент неуспевающим на данный момент. Список студентов чей процентиль последнего завершенного ими модуля 0.95 и более подаем продюсерам.

```python
# берем уже готовый датасет с завершенными модулями и имеющими домашнее задание,
#  с датами начала и окончания их выполнения по каждому студенту.
df_hw_done
```

```python
# копируем его в отдельный датасет
df_top_1 = df_hw_done.copy()
```

```python
# добавляем в датасет 21(максимальное количество модулей на всех курсах) колонку
# для занесения в них дат завершения по каждому модулю у каждого студента
it = list(itertools.repeat("finish",21))
for i, j in enumerate(it):
    it[i] =  it[i] + '_' + str(i+1)

df_top_1[it] = None
```

```python
df_top_1
```

```python
# переносим даты в нужные колонки
for x, y in enumerate(df_top_1.index):
    df_top_1.iloc[x, (df_top_1.iloc[x,2]+5)] = df_top_1.iloc[x,4]
df_top_1
```

```python
# удаляем ненужные столбцы и переводим столбцы с датами в формат datetime
df_top_1 = df_top_1.drop(['finish_date', 'module_end', 'module_number'], axis=1)
df_top_1.iloc[:, 3:] = df_top_1.iloc[:, 3:].apply(pd.to_datetime)
```

```python
# группируем датасет по студентам и получаем датасет,
# в котором в строке id студента, курс, начала первой домашней работы и даты завершения каждого модуля
df_top_2 = df_top_1.groupby('progress_id').agg(lambda x: x.min()).reset_index()
df_top_2
```

```python
# добавляем еще 21 колонку для подсчета времени прохождения курса до каджого модуля каждым студентом
times = list(itertools.repeat("time",21))
for i, j in enumerate(times):
    times[i] =  times[i] + '_' + str(i+1)

df_top_2[times] = None
```

```python
df_top_2
```

```python
# рассчитываем время по каждому модулю и округляем до количества полных дней
for x, y in enumerate(df_top_2.index):
    for i in range(21):
        df_top_2.iloc[x,i+24] = (((df_top_2.iloc[x,i+3] - df_top_2.iloc[x, 2]).days) + 1)
df_top_2
```

```python
# переводим новые колонки в цифровой формат
df_top_2[times] = df_top_2[times].apply(pd.to_numeric, errors='coerce')
```

```python
# копируем датасет и удаляем ненужные колонки
df_top_3 = df_top_2.copy()
df_top_3 = df_top_3.drop(df_top_3.iloc[:,3:24], axis=1)
```

```python
# еще раз добавляем 21 колонку для рассчета процентиля по каждому модулю на каждом курсе
pcts = list(itertools.repeat("pct",21))
for i, j in enumerate(times):
    pcts[i] =  pcts[i] + '_' + str(i+1)

df_top_3[pcts] = None
```

```python
# считаем процентиль и записываем в нужные колонки
for x in range(21):
    df_top_3[df_top_3.columns[x+24]] = df_top_3.groupby(['course_title'])[df_top_3.columns[x+3]].rank(pct=True)
df_top_3
```
